```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
from scipy.optimize import fsolve
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
x=np.linspace(100,499,801)
a=np.arange(0.01, 0.02, 0.0001)
```

```python
def equation(p):
    a,x=p
    return (np.sin(a*x), np.sin(a*x))

a1,x1=fsolve(equation, (0.015,250))
print(a1,x1)
```

```python
#Другой способ
a2=[]
x2=[]
for i in x:
    for j in a:
        if (np.sin(i*j)<1e-10):
            a2.append(i)
            x2.append(j)
```
