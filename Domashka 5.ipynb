```python
#3.
#1)	Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей
#(через биномиальное распределение) и сравните результаты.
print('На примере подбрасывания монетки')
print('События А - орел и В - решка несовместны, противоположные')
print('Примем, что вероятность выпадение орла = 0.5. Вероятность выпадение решки = 0.5. ')
print('Тогда:')
n=10000 #число экспериментов по подбрасыванию монетки
orel=reshka=0

for i in range (n):
    if np.random.randint(0,2)==0:
        orel=orel+1
    else:
        reshka=reshka+1
        
print(f'P(A+B)=P(A)+P(B)={orel/n}+{reshka/n}={orel/n+reshka/n}')
print(f'P(A)+P(B)={orel/n}+{reshka/n}={orel/n+reshka/n}')
print('Расчет вероятностей через биномиальное распределение:')
k,n=1,2
print(f'n = {n}, k = {k}')
Cnk=math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
print(f'Сочетания из n элементов по k: {Cnk}')
print(f'Вероятность k успехов среди n испытаний: {Cnk/(2**n)}')
```

```python
#4. Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k
#Пусть n=6, k=3
for p in itertools.permutations("012345", 3):
    print(''.join(str(x) for x in p))
```

```python
for p in itertools.combinations("012345", 3):
    print(''.join(p))
```

```python
for p in itertools.product("012345", repeat=6):
    print(''.join(p))
```
