```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 210)
k1=1
k2=4.5
y1=np.cos(k1*x)
y2=np.cos(k2*x)

plt.plot(x, y1)
plt.plot(x, y2)

plt.title("Графики функций")
plt.xlabel("Переменная x")
plt.ylabel("Функция f(k, x)")

plt.show()
```
